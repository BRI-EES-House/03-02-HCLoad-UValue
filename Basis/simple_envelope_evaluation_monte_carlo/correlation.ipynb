```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
df = pd.read_csv('db_env_spec.csv')
```

```python
upper_u_mean = [df[df['region'] == region]['upper_u'].mean() for region in [1,2,3,4,5,6,7,8]]
wall_u_mean = [df[df['region'] == region]['wall_u'].mean() for region in [1,2,3,4,5,6,7,8]]
d2 = np.mean([float(d) for d in df[(df['region'] == 1) & (df['lower_ins_type'] == '床断熱')]['lower_u']])
d2 = np.mean([float(d) for d in df[(df['region'] == 1) & (df['lower_ins_type'] == '床断熱')]['lower_u']])
d2 = df[df['lower_ins_type'] == '床断熱']['lower_u'].mean()
print(d2)
#lower_u_mean = [df[(df['region'] == region) & (df['lower_ins_type'] == '床断熱')]['lower_u'].mean() for region in [1,2,3,4,5,6,7,8]]

left = np.arange(8)
width = 0.1

fig = plt.figure(figsize=(15, 4))
ax = fig.add_subplot(1,1,1)
ax.bar(left, upper_u_mean, width=width, label='upper_u')
ax.bar(left+width, wall_u_mean, width=width, label='wall_u')
#ax.bar(left+width*2, lower_u_mean, width=width, label='lower_u')


ax.set_xticks([0,1,2,3,4,5,6,7])
ax.set_xticklabels(['region 1','region 2','region 3','region 4','region 5','region 6','region 7','region 8'])

plt.legend()
plt.show()
```
